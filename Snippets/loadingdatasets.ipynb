```python
p1 = "/home/asetia/remoteData/GLDAS_NEW/GLDAS_NOAH10_M/"
nldas = []  # List to store individual DataFrames

# Get the total number of NetCDF files
file_count = len([filename for filename in os.listdir(p1) if filename.endswith(".nc4")])

# Iterate over the files
for i, filename in enumerate(os.listdir(p1)):
    if filename.endswith(".nc4"):  # Select only NetCDF files
        file_path = os.path.join(p1, filename)
        data = xr.open_dataset(file_path)
        df = data.to_dataframe().reset_index()
        nldas.append(df)

        # Print progress
        print(f"Processed file {i+1} of {file_count}: {filename}")

# Concatenate individual DataFrames into a single DataFrame
merged_nldas = pd.concat(nldas, ignore_index=True)

# Now you have a merged DataFrame with all files in tabular format
print("Merge completed successfully.")
print(merged_nldas.head())  # Print a preview of the merged DataFrame


```
