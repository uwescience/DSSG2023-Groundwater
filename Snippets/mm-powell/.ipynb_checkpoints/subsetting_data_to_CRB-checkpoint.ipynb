## Loading GRACE data

```python
grace_path = '/home/maiampow/remoteData/GRACE/data/TELLUS_GRAC_L3_CSR_RL06_LND_v04/'
```

```python
def read_data_all(path, filename):
    
    if filename.endswith('nc'):
        df = xr.open_dataset(path+str(filename))
        d = df.to_dataframe()
        d.reset_index(inplace=True)
        data = d[["time","lat","lon","lwe_thickness"]]
        return data
    
    else:
        pass
```

```python
tdf = pd.DataFrame()

for filename in os.listdir(grace_path):
    temp_df = read_data_all(grace_path, filename)
    tdf = pd.concat([tdf,temp_df])
```

```python
grace_df_total = tdf[tdf.lwe_thickness.notnull()] #Removing rows with NaN values
```

```python
grace_df_total["lon"] = [k-180 for k in grace_df_total["lon"]] #converting longitude points
```

```python
grace_df_total["point"] = [Point(x,y) for x,y in zip(grace_df_total["lon"],grace_df_total["lat"])] #creating column of Point(lon,lat)
```

```python
#Keeping only rows with longitude and latitude in CRB

grace_final_df = pd.DataFrame()
itera = 0


for sh in s["geometry"]: #iterating through each region in the CRB 
    
    grace_df_total_2 = grace_df_total[(sh.contains(grace_df_total["point"]))]
    grace_final_df = pd.concat([grace_final_df,grace_df_total_2])
    
    print(itera)
    itera+=1
```

```python
grace_final_df = grace_final_df.sort_values(by=['time'])
grace_final_df.reset_index(inplace=True)
```

```python
grace_final_df
```

```python
#Averaging by date

avg_grace_data = pd.DataFrame()

for k in collections.Counter(grace_final_df["time"]).keys():
    k_df = pd.DataFrame({"time": [grace_final_df[grace_final_df["time"] == k].describe()["time"]["mean"]],
            "lwe_thickness": [grace_final_df[grace_final_df["time"] == k].describe()["lwe_thickness"]["mean"]]})
    avg_grace_data = pd.concat([avg_grace_data,k_df])
```

```python
avg_grace_data = avg_grace_data.sort_values(by=['time'])
avg_grace_data.reset_index(inplace=True)
```

```python
plt.figure(figsize=[10,8])
plt.plot(avg_grace_data["time"], avg_grace_data["lwe_thickness"])
plt.show()
```

```python
filenames = []
i = 0

for tp in collections.Counter(grace_final_df["time"]).keys():
    
    fig, ax = plt.subplots(figsize=(12, 6))
    s.plot(color="lightgrey", ax=ax)
    
    sub_df = grace_final_df[grace_final_df["time"] == tp]

    x = sub_df['lon']
    y = sub_df['lat']
    z = sub_df['lwe_thickness']
    plt.scatter(x, y, c=z,
                cmap='spring',marker="s", sizes=[500]*len(x))
    plt.colorbar(label='LWE')
    plt.clim(-0.18,0.21)
    plt.title(str(tp))
    
    # plot the line chart
    #plt.plot(y[:i])
    #plt.ylim(20,50)
    
    # create file name and append it to a list
    filename = f'{i}.png'
    i+=1
    filenames.append(filename)
    
    # save frame
    plt.savefig(filename)
    plt.close()
# build gif
with imageio.get_writer('lwe.gif', mode='I') as writer:
    for filename in filenames:
        image = imageio.imread(filename)
        writer.append_data(image)
        
# Remove files
for filename in set(filenames):
    os.remove(filename)
```
