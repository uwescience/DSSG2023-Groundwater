```python
import netCDF4 as nc
import pandas as pd
import os
import numpy as np
import geopandas as gpd
import matplotlib.pyplot as plt
import collections
import matplotlib.font_manager
import xarray as xr
import statistics
import geodatasets
import shapefile
from shapely.geometry import shape, Point
import imageio
from IPython.display import Image

plt.style.use('ggplot')
```

## Importing updated GRACE data (mascon)

```python
grace_path = '/home/maiampow/remoteData/GRACE_MASCON/data/TELLUS_GRAC-GRFO_MASCON_CRI_GRID_RL06.1_V3/'
```

```python
!ls /home/maiampow/remoteData/GRACE_MASCON/data/TELLUS_GRAC-GRFO_MASCON_CRI_GRID_RL06.1_V3
```

```python
df = xr.open_dataset(grace_path+"GRCTellus.JPL.200204_202304.GLO.RL06.1M.MSCNv03CRI.nc")
```

```python
d = df.to_dataframe()
d.reset_index(inplace=True)
data = d[["time","lat","lon","lwe_thickness","scale_factor"]]
```

```python
#Removing rows with NaN values
data2 = data[data.lwe_thickness.notnull()] 
grace_df = data2[data2.scale_factor.notnull()] 
```

```python
grace_df["lwe"] = [x*y for x,y in zip(grace_df["lwe_thickness"],grace_df["scale_factor"])] #multiplying scaling factor
```

```python
grace_df["lon"] = [k-180 for k in grace_df["lon"]] #converting longitude points
```

```python
grace_df["point"] = [Point(x,y) for x,y in zip(grace_df["lon"],grace_df["lat"])] #creating column of Point(lon,lat)
```

```python
grace_df
```

## Entire Colorado River Basin

```python
#Loading shapefile of CRB (comprising of 8 total regions, including Upper and Lower basin)
s = gpd.read_file("/home/maiampow/code/Colorado_River_Basin_Hydrological_Boundaries_with_Areas_served_by_Colorado_River.shx")


# plot of entire Colorado River Basin
plt.figure(figsize=[10,10])
s.plot()
plt.title("Colorado River Basin")
plt.show()
```

```python
#Keeping only rows with longitude and latitude in CRB

grace_final_df = pd.DataFrame()
itera = 0


for sh in s["geometry"]: #iterating through each region in the CRB 
    
    grace_df_total_2 = grace_df[(sh.contains(grace_df["point"]))]
    grace_final_df = pd.concat([grace_final_df,grace_df_total_2])
    
    print(itera)
    itera+=1
```

```python
crb_df = grace_final_df.sort_values(by=['time'])
crb_df.reset_index(inplace=True)
```

```python
crb_df
```

```python
crb_df.describe()
```

```python
filenames = []
i = 0

for tp in collections.Counter(crb_df["time"]).keys():
    
    fig, ax = plt.subplots(figsize=(12, 6))
    s.plot(color="lightgrey", ax=ax, edgecolor='k')
    
    sub_df = crb_df[crb_df["time"] == tp]

    x = sub_df['lon']
    y = sub_df['lat']
    z = sub_df['lwe']
    plt.scatter(x, y, c=z,
                cmap='spring_r',marker="s", sizes=[100]*len(x))
    plt.colorbar(label='LWE')
    plt.clim(-88,82)
    plt.title(str(tp)[0:10])
    
    # plot the line chart
    #plt.plot(y[:i])
    #plt.ylim(20,50)
    
    # create file name and append it to a list
    filename = f'{i}.png'
    i+=1
    filenames.append(filename)
    
    # save frame
    plt.savefig(filename)
    plt.close()
# build gif
with imageio.get_writer('updated_grace_crb.gif', mode='I') as writer:
    for filename in filenames:
        image = imageio.imread(filename)
        writer.append_data(image)
        
# Remove files
for filename in set(filenames):
    os.remove(filename)
```

```python
Image(url='figs/updated_grace_crb.gif')
```
