```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import geopandas as gpd
import shapefile
from shapely.geometry import shape, Point

plt.style.use('ggplot')
```

```python
s = gpd.read_file("Colorado_River_Basin_Hydrological_Boundaries_with_Areas_served_by_Colorado_River.shx")
```

```python
# plot of entire Colorado River Basin
plt.figure(figsize=[10,10])
s.plot()
plt.title("Colorado River Basin")
plt.show()
```

```python
# checking if specific point is within polygon (found on StackExchange)

def check(lon, lat, p):
    point = Point(lon, lat) # build a shapely point from your geopoint
    return p.contains(point)
```

```python
# visualizing each of the eight regions of the Colorado River Basin
for sh in s["geometry"]:
  gpd.GeoSeries(sh).plot()
plt.show()
```

```python
# extracting upper and lower basin from CRB
u_basin = s['geometry'][7]
l_basin = s['geometry'][6]
```

```python
gpd.GeoSeries(u_basin).plot()
plt.title("Upper Basin CRB")
plt.show()
```

```python
gpd.GeoSeries(l_basin).plot()
plt.title("Lower Basin CRB")
plt.show()
```

```python
# testing function to check if point of interest is within polygon (area of basin) of interest
print(check(-112,34,l_basin))
print(check(-116,36,l_basin))
```
