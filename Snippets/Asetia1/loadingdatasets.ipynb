```python
p1 = "/home/asetia/remoteData/GLDAS_NEW/GLDAS_NOAH10_M/"
nldas = []  # List to store individual DataFrames

# Get the total number of NetCDF files
file_count = len([filename for filename in os.listdir(p1) if filename.endswith(".nc4")])

# Iterate over the files
for i, filename in enumerate(os.listdir(p1)):
    if filename.endswith(".nc4"):  # Select only NetCDF files
        file_path = os.path.join(p1, filename)
        data = xr.open_dataset(file_path)
        
        columns_of_interest = ['time','lon', 'lat', 'SWE_inst','SoilMoi0_10cm_inst',
       'SoilMoi10_40cm_inst', 'SoilMoi40_100cm_inst', 'SoilMoi100_200cm_inst']
        
        df = data[columns_of_interest].to_dataframe().reset_index()
        nldas.append(df)
        

# Concatenate individual DataFrames into a single DataFrame
merged_nldas = pd.concat(nldas, ignore_index=True)

# Now you have a merged DataFrame with all files in tabular format
print("Merge completed successfully.")
print(merged_nldas.head())  # Print a preview of the merged DataFrame
```

```python
##Reading GRACE DATA

def read_data(path, filename, data):
    try:
        if filename.endswith('nc'):
            df = xr.open_dataset(path+str(filename))
            date = pd.to_datetime(df['time'].values[0])
            data['date'].append(date)
            region = df['lwe_thickness'].sel(lat=slice(30.5, 39.5), lon=slice(270.5, 279.5))
            data['lwe_thickness'].append(region.mean().values.item())
        else:
            pass
    except Exception as e:
        print(e)
        print(path)
        print(filename)
```

```python
grace_path = '/home/asetia/remoteData/GRACE_NEW/data/TELLUS_GRAC_L3_CSR_RL06_LND_v04/'
data = {'date':[], 'lwe_thickness':[]}

for filename in os.listdir(grace_path):
    read_data(grace_path, filename, data)
    df = pd.DataFrame(data)
df = df.sort_values('date')
df.dtypes
```
