```python
p1 = "/home/asetia/remoteData/GLDAS_NEW/GLDAS_NOAH10_M/"
nldas = []  # List to store individual DataFrames

# Get the total number of NetCDF files
file_count = len([filename for filename in os.listdir(p1) if filename.endswith(".nc4")])

# Iterate over the files
for i, filename in enumerate(os.listdir(p1)):
    if filename.endswith(".nc4"):  # Select only NetCDF files
        file_path = os.path.join(p1, filename)
        data = xr.open_dataset(file_path)
        
        columns_of_interest = ['time','lon', 'lat', 'SWE_inst','SoilMoi0_10cm_inst',
       'SoilMoi10_40cm_inst', 'SoilMoi40_100cm_inst', 'SoilMoi100_200cm_inst']
        
        df = data[columns_of_interest].to_dataframe().reset_index()
        nldas.append(df)
        
        print(f"Processed file {i+1} of {file_count}: {filename}")

# Concatenate individual DataFrames into a single DataFrame
merged_nldas = pd.concat(nldas, ignore_index=True)

# Now you have a merged DataFrame with all files in tabular format
print("Merge completed successfully.")
print(merged_nldas.head())  # Print a preview of the merged DataFrame


```

```python
#Estimates of the Coordinates of Mississippi: 47.5, 95.5
```

```python
Mississippi = merged_nldas[(merged_nldas['lat'] == 47.5) & (merged_nldas['lon'] == 95.5)]
```

```python
Mississippi
```

```python
Mississippi = Mississippi.sort_values('time')
Mississippi
```

```python
#Descriptives of the Data
Mississippi['time'].min() #1948-01-01
Mississippi['time'].max() #2014-12-01

#The data is monthly from 1948 till 2014
```

```python
plt.figure(figsize=(12, 6))
plt.plot(list(range(0,804)),Mississippi["SWE_inst"])
#Mississippi["SWE_inst"].plot()
plt.xlabel("time")
plt.ylabel("Snow Water Equivalent")
plt.title("Snow Water Equivalent in Mississippi River Basin")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(list(range(0,804)),Mississippi[['SoilMoi0_10cm_inst','SoilMoi10_40cm_inst', 'SoilMoi40_100cm_inst', 'SoilMoi100_200cm_inst']])
plt.xlabel('Time')
plt.ylabel('Soil Data')
plt.title('Time Series Map of Soil Data')
plt.grid(True)

custom_labels=['Soil Moisture 0 to 10 cm', 'Soil Moisture 10 to 40 cm', 'Soil Moisture 40 to 100 cm', 'Soil Moisture 100 to 200 cm']
plt.legend(custom_labels, loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.show()
```
